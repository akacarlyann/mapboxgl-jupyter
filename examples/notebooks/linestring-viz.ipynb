## Vector linestring source

```python
match_stops = [
    [0, "red"], 
    [10, "blue"], 
    [20, "green"], 
    [30, "yellow"], 
    [40, "pink"],
    [50, "magenta"],
    [60, "black"],
    [70, "cyan"],
    [80, "grey"],
    [90, "brown"],
    [200, "orange"],
]

viz = LinestringViz(data, 
                    vector_url='mapbox://mapbox.mapbox-terrain-v2',
                    vector_layer_name='contour',
                    vector_join_property='ele',
                    data_join_property='elevation',
                    color_property='elevation',
                    color_stops=create_color_stops([0, 25, 50, 75, 100], colors='YlOrRd'),
                    # color_stops=match_stops,
                    color_function_type='match',                    
                    line_width_property='something',
                    line_width_stops=create_radius_stops([0, 50, 100], 2, 6),
                    line_stroke='-',
                    line_width_default=3,
                    opacity=0.8,
                    center=(-122.48, 37.83),
                    zoom=16,
                    below_layer='waterway-label'
                   )
viz.show()
```
