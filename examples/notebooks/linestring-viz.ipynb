```python
import sys
import os
import pandas as pd

from mapboxgl.viz import *
from mapboxgl.utils import *

# Must be a public token, starting with `pk`
token = os.getenv('MAPBOX_ACCESS_TOKEN')

data = [{"elevation": 0, "something": 0},
        {"elevation": 10, "something": 200},
        {"elevation": 20, "something": 200},
        {"elevation": 30, "something": 30},
        {"elevation": 40, "something": 200},
        {"elevation": 50, "something": 20},
        {"elevation": 60, "something": 200},
        {"elevation": 70, "something": 40},
        {"elevation": 80, "something": 200},
        {"elevation": 90, "something": 200},
        {"elevation": 100, "something": 200},
       ]
# https://www.mapbox.com/mapbox-gl-js/example/vector-source/

match_stops = [
    [0, "red"], 
    [10, "blue"], 
    [20, "green"], 
    [30, "yellow"], 
    [40, "pink"],
    [50, "magenta"],
    [60, "black"],
    [70, "cyan"],
    [80, "grey"],
    [90, "brown"],
    [200, "orange"],
]

viz = LinestringViz(data, 
                    vector_url='mapbox://mapbox.mapbox-terrain-v2',
                    vector_layer_name='contour',
                    vector_join_property='ele',
                    data_join_property='elevation',
                    color_property='elevation',
                    color_stops=match_stops,
                    color_function_type='match',
                    color_default='rgb(255,255,0)',
                    line_stroke='-',
                    line_color='notused', ##########
                    line_width=1,
                    opacity=0.8,
                    center=(-122.48, 37.83),
                    zoom=16,
                    below_layer='waterway-label'
                   )
viz.show()

```
